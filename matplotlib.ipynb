#### VERİ GÖRSELLEŞTİRME (MATPLOTLİB)

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
x=["1.ABD","2.ÇİN","3.JAPONYA","4.ALMANYA","5.İNGİLTERE","6.FRANSA"]
y=[20410,14090,5170,4210,2940,2930]
z=[63039,9957,40649,51158,44162,44915]
```

```python
plt.plot(x,y,"r--")
```

```python
plt.plot(x,y,"g-*")
plt.xlabel("ÜLKELER")
plt.ylabel("MİLLİ GELİR")
plt.title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.plot(x,z,"g-*")
plt.tight_layout()
```

```python
f=plt.figure(figsize=(8,3))
axes=f.add_axes([0.3,0.3,0.9,0.9])
axes.plot(x,y,color="blue",linewidth=8,linestyle="--",marker="o",markerfacecolor="yellow",markeredgecolor="red")
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
f.savefig("resim.png") #kaydettim 
```

```python
fig=plt.figure(figsize=(10,5))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,"r",label="MİLLİ GELİR")
ax.plot(x,z,"y",label="KİŞİ BAŞINA")
ax.legend()
```

```python
plt.style.use('_mpl-gallery')
# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])
```
