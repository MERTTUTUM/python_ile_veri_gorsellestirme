```python
n=pd.read_csv("C:/Users/Mertcan/Desktop/python kurs\KAYNAK/seaborn/dünyam.csv")
n
```

```python
sns.set(style="darkgrid",palette="muted",font_scale=1.5)
sns.displot(n["Doğum oranı"],color="m",bins=20)
```

```python
sns.set(style="whitegrid",palette="Blues",font_scale=1.5)
sns.displot(n["Nüfüs"],color="r")
```

```python
sns.set(style="darkgrid",font_scale=1)
f,axes=plt.subplots(2,2,figsize=(15,10))

sns.distplot(n["Doğum oranı"],color="m",bins=20,hist=False,ax=axes[0,0])
sns.distplot(n["Yıllık değişim oranı"],color="r",ax=axes[0,1])
sns.displot(n["Ortalama yaş"],color="g",kde_kws={"shade":True},ax=axes[1,0])
sns.displot(n["Nüfüs"],color="y",bins=30,ax=axes[1,1])
```

###### jointplot()

```python
sns.jointplot(n["Doğum oranı"],n["Ortalama yaş"],data=n)
```

```python
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n,kind="hex",color="m")
```

```python
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n,kind="kde",color="g")
```
